```python
df=pd.read_excel('/content/gdrive/MyDrive/all_academic_necessities/INTERN/official_work/aqualink/HR_data/report_jan21_jan22.xlsx')
df.head()
```

```python
# removing the rows which contains friday and saturday

dropday = df[ (df['Week'] == 'Friday') | (df['Week'] == 'Saturday') ].index
df.drop(dropday , inplace=True)
df.head()
```

```python
 #converting the string date to timestamp
 
 df['Office_Duration'] = pd.to_timedelta(df['Office_Duration'])
 df
```

```python
# this is how we can access different time comoponent

df.Office_Duration.dt.components.hours
```

```python
#df['attendance_type'] = df['Office_Duration'].apply(lambda x: 'P' if df.Office_Duration.dt.components.hours >= 4 else 'A')
```

```python
# condition for attendance type

df.loc[df.Office_Duration.dt.components.hours >= 4, 'attendance_type'] = 'P'
df.loc[(df.Office_Duration.dt.components.hours < 4) | (df.Office_Duration.isnull()==True), 'attendance_type'] = 'A'


```

```python
# for some reason there was a mistake that caused office type not null where absent , that's why this...

df.loc[(df['attendance_type']=='A'), 'office_type'] = np.nan
```

**we can count the number of values in a pandas DataFrame column with a specific condition:**

**Method 1: Count Values in One Column with Condition**

**len(df[df['col1']=='value1'])**

**Method 2: Count Values in Multiple Columns with Conditions**

**len(df[(df['col1']=='value1') & (df['col2']=='value2')])**
